A cada dia novos casos seriam identificados, desses casos era possivel acompanhar os casos de recuperados,dos que vieram a obitos e os que aindam estavam em recuperação. Diante deste cenário atual queiriamos trabalhar e tratar esses dataset usando o pandas no jupyter notebook como ferramenta para monitorar esses casos. 

Aqui iremos usar o dataset disponiblizado no http://plataforma.saude.gov.br/coronavirus/covid-19/ 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('summary.csv', sep=';')
df.head()
```

Aqui iremos fazer o tratamento dos dados para melhor calcular, vamos atribuir 0 para cada dados faltantes nas colunas do dataset.

```python
df['Casos Confirmados - Acumulado'].fillna(0).implace = True
df['Casos Confirmados - Novos'].fillna(0).implace = True
df['Casos Confirmados - Incremento'].fillna(0).implace = True
df['Casos Confirmados - Incidência*'].fillna(0).implace = True
df['Casos Confirmados - Em acompanhamento'].fillna(0).implace = True
df['Casos Confirmados - Recuperados'].fillna(0).implace = True
df['Óbitos Confirmados - Acumulado'].fillna(0).implace = True
df['Óbitos Confirmados - Novos'].fillna(0).implace = True
df['Óbitos Confirmados - Incremento'].fillna(0).implace = True
df['Óbitos Confirmados - Mortalidade*'].fillna(0).implace = True
df['Óbitos Confirmados - Letalidade'].fillna(0).implace = True
```

```python
df.hist(column='Casos Confirmados - Novos', bins=100)
plt.show()
```
